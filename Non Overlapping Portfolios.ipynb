```python
import yfinance as yf
import numpy as np
import pandas as pd
from IPython.display import clear_output
from pandas.tseries.offsets import MonthEnd
```

```python
stocks = pd.read_html('https://en.wikipedia.org/wiki/NIFTY_50')[2].Symbol 
more_stocks = pd.read_html('https://en.wikipedia.org/wiki/NIFTY_50')[3].Symbol
more_more_stocks = pd.read_html('https://en.wikipedia.org/wiki/NIFTY_50')[3].Symbol_2 #obtains all stocks that have been in nifty since 2010
```

```python
stocks += '.NS'
more_stocks += '.NS'
more_more_stocks += '.NS'
```

```python
stocks = stocks.to_list()
more_stocks = more_stocks.to_list()
more_more_stocks = more_more_stocks.to_list()

stocks = stocks + more_stocks + more_more_stocks
stocks_nd = []
[stocks_nd.append(x) for x in stocks if x not in stocks_nd]

stocks = stocks_nd
```

```python
df = yf.download(stocks,start='2010-01-01')['Close'] # downloads the daily closing price of all the Nifty-50 sonsituent stocks from 1st January,2010 until present
```

```python
df
```

```python
ret_df = df.pct_change() #gets the percentage change per day
```

```python
ret_df
```

```python
ret_df.index = pd.to_datetime(ret_df.index)
mtl_ret = (ret_df + 1).resample('M').prod() #converts the data to returns for the month (relative to price at the start of the month)
```

```python
mtl_ret
```

```python
def top_performers(date,lookback_period,holding_period,n):
    try:
        mtl_lb = mtl_ret.rolling(lookback_period).apply(np.prod).dropna() #cumulative movement in the stock price over the previous "lookback_period" months
        top = mtl_lb.loc[date].nlargest(n) #gets the n best performing stocks for the given 'lookback_period' months 
        relevant_ret = mtl_ret[top.name:][1:1+holding_period][top.index].rolling(holding_period).apply(np.prod)
        return (relevant_ret).mean(axis = 1).values[holding_period-1] #returns the average returns for the top 'n' stocks given a holding period
    except KeyError:
        return 1
```

```python
def getPortfolioReturns(lookback_period,holding_period,n):
    
    momentum_returns = []
    i=0
    for date in mtl_ret.index[:114:holding_period]: 
        momentum_returns.append(top_performers(date,lookback_period,holding_period,n))
        clear_output(wait=True)
        print("[   Datapoint: %0.2f percent done   ]" % (i*holding_period/115*100))
        i+=1
        
    clear_output(wait=True)
    return round(pd.Series(momentum_returns).prod(),2)
          
```

```python
#compilation of results from different Momentum Strategies

returns = []
for n in [5,10,15,20,25]:
    returns_for_given_n = []
    for J in [3,6,12]:
        returns_for_given_n_row = []
        
        for K in [1,3,6,12]:
            returns_for_given_n_row.append(getPortfolioReturns(J,K,n))
            
        returns_for_given_n.append(returns_for_given_n_row)
        
    returns.append(returns_for_given_n) 

clear_output(wait=True)
print(returns)
```

```python
#Incorporating a one month lag 

def top_performers_lagged(date,lookback_period,holding_period,n):
    try:
        prev_date = date - MonthEnd(1)
        mtl_lb = mtl_ret.rolling(lookback_period-1).apply(np.prod).dropna() #cumulative movement in the stock price over the previous "lookback_period" months
        top = mtl_lb.loc[prev_date].nlargest(n) #gets the n best performing stocks for the given 'lookback_period' months 
        relevant_ret = mtl_ret[top.name:][2:2+holding_period][top.index].rolling(holding_period).apply(np.prod) #changed 1 to 2 
        return (relevant_ret).mean(axis = 1).values[holding_period-1] #returns the average returns for the top 'n' stocks given a holding period
    except KeyError:
        return 1
    
def getPortfolioReturns_lagged(lookback_period,holding_period,n):
    
    momentum_returns = []
    i=0
    for date in mtl_ret.index[1:114:holding_period]: 
        momentum_returns.append(top_performers_lagged(date,lookback_period,holding_period,n))
        clear_output(wait=True)
        print("[   Datapoint: %0.2f percent done   ]" % (i*holding_period/115*100))
        i+=1
        
    clear_output(wait=True)
        
    return round(pd.Series(momentum_returns).prod(),2)
          
```

```python
#compilation of results from different Momentum Strategies

returns2 = []
for n in [5,10,15,20,25]:
    returns_for_given_n = []
    for J in [3,6,12]:
        returns_for_given_n_row = []
        
        for K in [1,3,6,12]:
            returns_for_given_n_row.append(getPortfolioReturns_lagged(J,K,n))
            
        returns_for_given_n.append(returns_for_given_n_row)
        
    returns2.append(returns_for_given_n) 

clear_output(wait=True)
print(returns2)
```
