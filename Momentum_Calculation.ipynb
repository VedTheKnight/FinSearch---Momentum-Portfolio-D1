```python
import yfinance as yf
import pandas as pd
import numpy as np 
```

```python
link = 'https://en.wikipedia.org/wiki/NIFTY_50'

symbols = pd.read_html(link)[2]['Symbol']

symbols = symbols + ".NS"

symbols = symbols.to_list()
```

```python
start = '2010-01-01'

prices = yf.download(symbols, start = start)['Close']

prices
```

```python
percent_change = prices.pct_change()

# Taking product of the daily fractional change over a month to get the fractional change in stock price over a month
monthly_return = (percent_change + 1).resample('M').prod()
```

```python
monthly_return
```

```python
def WinnerStocks(date, lookback, holding, no_stocks):
    try:
        # Define the monthly returns in stocks over the lookback period
        monthly_return_lb = monthly_return.rolling(lookback - 1).apply(np.prod).dropna()

        # Pick the largest n stocks from the lookback period given a date
        winners = monthly_return_lb.loc[date - pd.offsets.MonthEnd(1)].nlargest(no_stocks)

        # Pick these stocks to invest in at the current date
        current_ret = monthly_return[winners.name:][1 : 1 + holding][winners.index].rolling(holding).apply(np.prod)

        # Return the average returns over the perios for all the stocks
        if not current_ret.empty:
            return (current_ret).mean(axis = 1).values[0]
    except:
        return 1
```

```python
def Portfolio_Returns(lookback, holding, no_stocks):
    port_returns = []

    # Define the monthly returns in stocks over the lookback period
    monthly_return_lb = monthly_return.rolling(lookback - 1).apply(np.prod).dropna()

    # Calculate the returns over the entire monthly return dataset
    for date in monthly_return_lb.index[:114:holding]:
        port_returns.append(WinnerStocks(date, lookback, holding, no_stocks))

    # Return the net returns of the investment scheme
    return round(pd.Series(port_returns).prod(), 10)
```

```python
Portfolio_Returns(3, 1, 5)
```

```python
# Compilation of Results from different Momentum Strategies

returns = []
for n in [5, 10, 15, 20]:
    returns_for_given_n = []
    for J in [3,6,9,12]:
        returns_for_given_n_row = []
        
        for K in [3,6,9,12]:
            returns_for_given_n_row.append(Portfolio_Returns(J,K,n))
            
        returns_for_given_n.append(returns_for_given_n_row)
        
    returns.append(returns_for_given_n) 

print(returns)
```
